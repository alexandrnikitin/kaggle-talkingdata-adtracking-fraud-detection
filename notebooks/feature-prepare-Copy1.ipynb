```python
dtypes = {
    'ip': 'uint32',
    'app': 'uint16',
    'device': 'uint16',
    'os': 'uint16',
    'channel': 'uint16',
    'is_attributed': 'uint8'
}
to_read = ['app', 'device', 'os', 'channel', 'is_attributed']

target_entities = ['ip', 'app', 'device', 'os', 'channel']
```

```python
train_files = [
 '../data/interim/train_2017-11-09_0400.csv',
 '../data/interim/train_2017-11-09_0500.csv',
 '../data/interim/train_2017-11-09_0600.csv'
]
```

```python
for train_file in train_files:
    df_train = pd.read_csv(train_files[0], usecols=to_read, dtype=dtypes)
    name, ext = os.path.splitext(os.path.basename(train_file))
    for target_entity in target_entities:
        feature_files = sorted(glob(f"../data/interim/features/{name}_{target_entity}_*_features.csv"))
        for feature_file in feature_files:
            df_feature = pd.read_csv(feature_file)
            df_feature = df_feature.set_index(target_entity)
            df_train = df_train.join(df_feature, on=target_entity)
    print(f"Writing ../data/interim/combined/{name}{ext}")
    df_train.to_csv(f"../data/interim/combined/{name}{ext}", index=False)
    del df_train
    gc.collect()
```
